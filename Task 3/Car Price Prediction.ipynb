```python
df['company'] = df['CarName'].apply(lambda x: x.split(' ')[0])
```

```python
df.company.unique()
```

```python
df.company = df.company.str.lower()
```

```python
name = {
    'toyouta': 'toyota',
    'vokswagen': 'volkswagen',
    'vw': 'volkswagen',
    'porcshce': 'porsche',
    'maxda': 'mazda'
}

df['company'] = df['company'].apply(lambda x: name[x] if x in name.keys() else x)
```

```python
df.company.unique()
```

```python
df.head()
```

```python
cars_lr = df[['price', 'fueltype', 'aspiration', 'carbody', 'drivewheel', 'wheelbase', 'curbweight', 'enginetype', 'cylindernumber', 'enginesize', 'boreratio','horsepower', 'carlength','carwidth']]
cars_lr.head()
```

```python
cols = cars_lr.select_dtypes('object').columns
cars_lr = pd.get_dummies(cars_lr, columns=cols, drop_first=True)
```

```python
len(cars_lr.columns)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X = cars_lr.drop('price', axis=1)
y = cars_lr['price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
lin = LinearRegression()
lin.fit(X_train, y_train)
lin.score(X_test, y_test)
```

```python
from sklearn.tree import DecisionTreeRegressor

dec = DecisionTreeRegressor()
dec.fit(X_train, y_train)
dec.score(X_test, y_test)
```
